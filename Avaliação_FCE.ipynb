**1) Considere a seguinte afirmação: “Programar não é apenas escrever código, mas estruturar logicamente uma solução.” Explique essa afirmação com base nos conceitos de algoritmo e lógica computacional. Em

O conceito de programar está diretamente relacionado ao pensamento lógico, uma vez que, diante de um problema, o jeito mais fácil de resolvê-lo é dividi-lo em pequenos passos até atingir a solução desejada. A programação funciona da mesma forma: depois de pensar em uma solução, define-se passo a passo um caminho para que seja possível chegar ao objetivo.
```python
# Programa para verificar se o número é impar ou par e se a entrada é numérica

entrada = input("Digite um número: ")

try:
    numero = int(entrada)

    if numero % 2 == 0:
        print("O número é par")
    else:
        print("O número é ímpar")

except ValueError:
    print("Digite apenas números e números inteiros")
```

(b) fluxograma correspondente utilizando o módulo graphviz em Python.

```python
# Fluxograma utilizando o graphviz
from graphviz import Digraph

# Cria um objeto grafo vazio
g = Digraph()

# =========================
# Definição dos nós (blocos)
# =========================

g.node('A', 'Início')
g.node('B', 'Ler nível')
g.node('C', 'Nível < 20%?')
g.node('D', 'Exibir: Nível baixo\ncontador +1')
g.node('E', 'Nível ≤ 80%?')
g.node('F', 'Exibir: Operação normal')
g.node('G', 'Exibir: Risco de transbordamento')
g.node('H', 'contador = 3')
g.node('I', 'Exibir: Alerta de parada')
g.node('J', 'Fim')

# =========================
# Definição das Conexões
# =========================

g.edge('A', 'B')
g.edge('B', 'C')

# Nível < 20%
g.edge('C', 'D', label='Sim')
g.edge('D', 'H')

# Caso contrário
g.edge('C', 'E', label='Não')

# Nível ≤ 80%
g.edge('E', 'F', label='Sim')
g.edge('F', 'H')

# Nível > 80%
g.edge('E', 'G', label='Não')
g.edge('G', 'H')

# Verificação do contador
g.edge('H', 'I', label='Sim')
g.edge('I', 'J')  # FIM- Devido a Parada

g.edge('H', 'B', label='Não')  # volta para o loop

# =========================
# Exibição
# =========================

g
```

(c) Código definindo as etapaas do processo:

```python
def monitorar_tanque():
    NIVEL_MIN = 20
    NIVEL_MAX = 80
    contador_baixo = 0

    while True:
        entrada = input("Digite o nível do tanque (0 a 100) ou 'sair': ")

        if entrada.lower() == 'sair':
            print("Encerrando monitoramento.")
            break

#  Permite que o usuario entre com uma medida ou encerre o monitoramento

        try:
            nivel = float(entrada)

            if nivel < 0 or nivel > 100:
                raise ValueError("O nível deve estar entre 0 e 100.")

# Verificando se a entrada atende o range esperado

            if nivel < NIVEL_MIN:
                print("Nível baixo")
                contador_baixo += 1

            elif nivel <= NIVEL_MAX:
                print("Operação normal")
                contador_baixo = 0

            else:
                print("Risco de transbordamento")
                contador_baixo = 0

            if contador_baixo == 3:
                print("ALERTA: Parada do sistema!")
                break

  # Caso a medida de nível baixo atinge 3 medições o programa encerra

        except ValueError as e:
            print(f"Erro: {e}")

monitorar_tanque()
```

**4) Durante a operação de um processo, a concentração de um reagente é monitorada periodicamente para avaliar as condições de operação do sistema. A concentração pode variar ao longo do tempo e deve ser classificada em diferentes faixas para auxiliar na análise do processo. Neste contexto, você recebeu a tarefa de criar um algoritmo que permita executar a simulação do sistema de monitoramento, o qual deve seguir os seguintes requisitos:**

**• A cada leitura, deve ser gerado um valor aleatório de concentração entre 0 e 10 mol/L.**

**• Se a concentração for menor que 3, o sistema deve indicar “Baixa”.**

**• Se a concentração estiver entre 3 e 7, o sistema deve indicar “Adequada”.**

**• Se a concentração for maior que 7, o sistema deve indicar “Alta”.**

**• Ao final da simulação, deve-se contabilizar quantas vezes cada condição ocorreu.**

**Com base nesses requisitos, desenvolva um código modularizado em Python para simular o monitoramento considerando 10 leituras consecutivas.
O código deve conter, obrigatoriamente:**

**• Pelo menos uma função**

**• Estrutura de repetição**

**• Estrutura condicional**

**• Comentários explicativos**

**Ao final da execução, o programa deve apresentar:**

**• O número de ocorrências de cada classe (Baixa, Adequada e Alta)**

**• Uma mensagem indicando qual condição foi predominante no processo**

**Um exemplo de informações ao usuário pode ser visto no Quadro 1.**



image.png


O código desenvolvido está disposto abaixo:

```python
import random

def classificar_concentracao(concentracao):
    if concentracao < 3:
        return "Baixa"
    elif concentracao <= 7:
        return "Adequada"
    else:
        return "Alta"

def monitorar_concentracao():
    baixa = 0
    adequada = 0
    alta = 0

    for i in range(1, 11):
        concentracao = round(random.uniform(0, 10), 1)
        classificacao = classificar_concentracao(concentracao)

        print(f"Leitura {i}: Concentração = {concentracao} mol/L -> {classificacao}")

        if classificacao == "Baixa":
            baixa += 1
        elif classificacao == "Adequada":
            adequada += 1
        else:
            alta += 1

    print("\nResumo:")
    print(f"Baixa: {baixa} ocorrências")
    print(f"Adequada: {adequada} ocorrências")
    print(f"Alta: {alta} ocorrências")

    if baixa > adequada and baixa > alta:
        predominante = "Baixa"
    elif adequada > baixa and adequada > alta:
        predominante = "Adequada"
    elif alta > baixa and alta > adequada:
        predominante = "Alta"
    else:
        predominante = "Empate entre condições"

    print(f"\nCondição predominante: {predominante}")


monitorar_concentracao()
```

**1)Considere a seguinte afirmação: “Programar não é apenas escrever código, mas
estruturar logicamente uma solução.” Explique essa afirmação com base nos conceitos de algoritmo e lógica computacional. Em
sua resposta, inclua um exemplo simples.**

O conceito de programar está diretamente relacionado ao pensamento lógico, uma vez que, diante de um problema, o jeito mais fácil de resolvê-lo é dividi-lo em pequenos passos até atingir a solução desejada. A programação funciona da mesma forma: depois de pensar em uma solução, define-se passo a passo um caminho para que seja possível chegar ao objetivo. Um exemplo deste processo é a formulação do pseudocódigo, que tem por objetivo estruturar os pensamentos lógicos que definem a solução do problema e a implementação da linha de pensamento na construção do código. Um exemplo pode ser encontrado no código abaixo, em que se utilizou o pensamento lógico para definir se o número é par ou impar e se a entrada é um número.

```python
# Programa para verificar se o número é impar ou par e se a entrada é numérica

entrada = input("Digite um número: ")

try:
    numero = int(entrada)

    if numero % 2 == 0:
        print("O número é par")
    else:
        print("O número é ímpar")

except ValueError:
    print("Digite apenas números e números inteiros")
```

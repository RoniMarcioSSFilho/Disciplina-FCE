```python
def geometria_quadrado(medida=10):
  '''
  Está função calcula a área
  e o perímetro de um quadrado.
  '''
  # Comandos da função
  try:
    if not medida>0:
      raise ValueError("A medida do lado deve ser maior que zero.")
    area = medida ** 2
    perimetro = medida

  except ValueError as e:
    print(f"Erro: {e}")
    area,perimetro= None, None

  return area, perimetro



```

```python
 resultado_area, resultado_perimetro = geometria_quadrado()
 print(f"Área: {resultado_area}")
 print(f"Perímetro: {resultado_perimetro}")
```

```python
from typing import Tuple
import math

def raizes_eq2grau(a:float, b: float, c: float) -> Tuple[str, complex, complex]:
  delta = b**2 -4*a*c

  if delta<0:
    # Retorna raízes complexas
    tipo='Raízes complexas'
    raiz1=complex(-b,math.sqrt(-delta)/(2*a))
    raiz2=complex(-b,-math.sqrt(-delta)/(2*a))
  else:
    #Retorna raízes reais
    tipo='Raízes reais'
    raiz1=complex((-b+math.sqrt(delta))/(2*a))
    raiz2=complex((-b-math.sqrt(delta))/(2*a))

  return tipo, raiz1, raiz2

print(raizes_eq2grau(1,2,-3))
```
